````python
# --------------------------
# æ°´å°å·¥å…·å‡½æ•¸å’Œå‡½å¼åº«å°å…¥
# --------------------------
import numpy as np
import hashlib
from scipy.stats import norm # åƒ…ç”¨æ–¼æª¢æ¸¬ï¼Œä½†ç‚ºå®Œæ•´æ€§ä¹Ÿä¿ç•™

def hash_token(token, seed=0):
    """å°‡ token è½‰æ›æˆå½éš¨æ©Ÿç¨®å­"""
    h = hashlib.sha256((str(token) + str(seed)).encode()).hexdigest()
    return int(h, 16)

def partition_vocab(vocab_size, seed, gamma=0.5):
    """å°‡è©è¡¨åˆ†æˆ green list å’Œ red list"""
    rng = np.random.default_rng(seed)
    perm = rng.permutation(vocab_size)
    split = int(gamma * vocab_size)
    # ä½¿ç”¨ Python set é€²è¡Œé«˜æ•ˆæŸ¥æ‰¾
    green = set(perm[:split].tolist())
    red = set(perm[split:].tolist())
    return green, red

# --------------------------
# æ°´å°æ¡æ¨£å™¨
# --------------------------
def watermark_sampling(logits, prev_token, gamma=0.5, delta=2.0, hard=False):
    """æ ¹æ“šæ°´å°è¦å‰‡ä¿®æ”¹ logits"""
    vocab_size = logits.shape[-1]
    # ä½¿ç”¨ 32 ä½æ•´æ•¸ä½œç‚ºç¨®å­ï¼Œèˆ‡æ‚¨çš„å·¥å…·å‡½å¼ä¸€è‡´
    seed = hash_token(prev_token) % (2**32)
    green, red = partition_vocab(vocab_size, seed, gamma)

    # å°‡ logits è½‰ç§»åˆ° CPU é€²è¡Œ numpy æ“ä½œï¼Œç„¶å¾Œå†å›å‚³ GPU (å¦‚æœéœ€è¦åœ¨ GPU ä¸Šé€²è¡Œæ¡æ¨£ï¼Œå‰‡éœ€èª¿æ•´)
    # ç”±æ–¼ Unsloth/transformers æ¨¡å‹çš„è¼¸å‡ºé€šå¸¸åœ¨ GPU ä¸Šï¼Œæˆ‘å€‘ç›´æ¥åœ¨ GPU ä¸Šé€²è¡Œæ“ä½œä»¥é¿å…é »ç¹çš„ CPU-GPU å‚³è¼¸ã€‚

    # æ³¨æ„ï¼šGreen/Red List æ‡‰è©²æ˜¯æ•´æ•¸åˆ—è¡¨æˆ–é›†åˆ
    green_indices = list(green)

    if hard:
        # ç¡¬ç´…åå–®ï¼šç›´æ¥å±è”½ç´…è‰²é›†åˆ
        # ç”±æ–¼ Unsloth çš„æ¨¡å‹æ˜¯ PyTorch tensorï¼Œä½¿ç”¨ torch.tensor é€²è¡Œç´¢å¼•å’Œæ“ä½œ
        mask = torch.full_like(logits, float("-inf"), device=logits.device)
        mask[green_indices] = 0
        logits = logits + mask
    else:
        # è»Ÿç´…åå–®ï¼šçµ¦ç¶ è‰²é›†åˆåŠ  Î´ åç½®
        bias = torch.zeros_like(logits, device=logits.device)
        bias[green_indices] = delta
        logits = logits + bias

    return logits

# --------------------------
# å¸¶æ°´å°çš„æ–‡æœ¬ç”Ÿæˆ
# --------------------------
def generate_with_watermark_unsloth(model, tokenizer, messages, max_new_tokens=64, gamma=0.5, delta=2.0, hard=False, temperature=1.0, top_p=0.95, top_k=64):
    device = next(model.parameters()).device # ç²å–æ¨¡å‹æ‰€åœ¨è¨­å‚™

    # æ‡‰ç”¨ chat template
    text = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt = True, # å¿…é ˆç‚º True
    )
    input_ids = tokenizer([text], return_tensors = "pt").input_ids.to(device)

    # æº–å‚™ for è¿´åœˆ
    generated_ids = input_ids.clone()

    # ç¢ºä¿æ¨¡å‹è™•æ–¼è©•ä¼°æ¨¡å¼
    model.eval()

    with torch.no_grad():
        for _ in range(max_new_tokens):
            # ç²å–ä¸‹ä¸€å€‹ token çš„ logits
            outputs = model(generated_ids)
            logits = outputs.logits[:, -1, :].squeeze(0) # ç²å–æœ€å¾Œä¸€å€‹ token çš„ logits

            # ç²å–å‰ä¸€å€‹ token
            prev_token = int(generated_ids[0, -1])

            # æ‡‰ç”¨æ°´å°æ¡æ¨£
            watermarked_logits = watermark_sampling(logits, prev_token, gamma=gamma, delta=delta, hard=hard)

            # æ‡‰ç”¨ Top-k/Top-p/Temperature æ¡æ¨£ (å¯é¸ï¼Œä½†ä¿æŒä¸€è‡´æ€§)
            # ç”±æ–¼æ°´å°å·²ç¶“ä¿®æ”¹äº† logitsï¼Œæˆ‘å€‘å¯ä»¥é¸æ“‡æ›´ç°¡å–®çš„æ¡æ¨£æ–¹å¼ï¼Œæˆ–è€…å…ˆæ‡‰ç”¨æ°´å°å†æ‡‰ç”¨ Top-k/Top-p
            # é€™è£¡æˆ‘å€‘ä½¿ç”¨ multinomial é€²è¡Œç°¡å–®æ¡æ¨£ï¼Œæ‚¨å¯ä»¥æ ¹æ“šéœ€è¦æ›´æ›ç‚º top-k/top-p é‚è¼¯

            # ä½¿ç”¨ softmax è½‰æ›ç‚ºæ©Ÿç‡
            probs = torch.softmax(watermarked_logits, dim=-1)

            # æ‡‰ç”¨ top-k/top-p ç¯©é¸ (å¦‚æœéœ€è¦æ›´ç²¾ç¢ºçš„æ¡æ¨£æ§åˆ¶ï¼Œå¯ä½¿ç”¨ `transformers.top_k_top_p_filtering`)
            # ç‚ºäº†ç°¡æ½”ï¼Œé€™è£¡ç›´æ¥ä½¿ç”¨ torch.multinomial é€²è¡Œæ¡æ¨£
            next_token = torch.multinomial(probs, num_samples=1).unsqueeze(0).to(device)

            # æª¢æŸ¥æ˜¯å¦ç‚º EOS tokenï¼Œå¦‚æœæ˜¯å‰‡åœæ­¢
            if next_token.item() == tokenizer.eos_token_id:
                break

            # å°‡æ–° token æ·»åŠ åˆ°å·²ç”Ÿæˆçš„åºåˆ—
            generated_ids = torch.cat([generated_ids, next_token], dim=1)

    # åªè§£ç¢¼ç”Ÿæˆçš„éŸ¿æ‡‰éƒ¨åˆ†ï¼ˆå»é™¤ promptï¼‰
    full_output = tokenizer.decode(generated_ids[0].cpu(), skip_special_tokens=False)
    # æ‰¾åˆ° '<start_of_turn>model\n' ä¹‹å¾Œçš„æ–‡æœ¬ä½œç‚ºéŸ¿æ‡‰
    response_start = full_output.rfind('<start_of_turn>model\n')
    if response_start != -1:
        # + len('<start_of_turn>model\n') è·³éæ¨™ç±¤
        response = full_output[response_start + len('<start_of_turn>model\n'):]
    else:
        response = full_output

    # æ¸…ç†æ‰å¯èƒ½çš„çµæŸæ¨™ç±¤
    response = response.replace('<end_of_turn>', '').strip()

    return response

# --------------------------
# æ°´å°æª¢æ¸¬ (å¯é¸ï¼Œç”¨æ–¼é©—è­‰)
# --------------------------
def detect_watermark(text, tokenizer, gamma=0.5):
    # é€™è£¡å‡è¨­è¼¸å…¥çš„ text æ˜¯ä¹¾æ·¨çš„éŸ¿æ‡‰æ–‡æœ¬
    tokens = tokenizer(text, return_tensors="pt", add_special_tokens=False).input_ids[0].tolist()
    if not tokens:
        return 0, 1.0 # å¦‚æœæ²’æœ‰ tokenï¼Œè¿”å›é»˜èªå€¼

    T = len(tokens) - 1 # æª¢æ¸¬å¾ç¬¬äºŒå€‹ token é–‹å§‹ï¼ŒT æ˜¯å¯æª¢æ¸¬çš„ token æ•¸é‡
    if T <= 0:
        return 0, 1.0

    green_count = 0

    for i in range(1, len(tokens)):
        # å‰ä¸€å€‹ token ç”¨æ–¼ç”Ÿæˆç¨®å­
        prev_token = tokens[i - 1]
        seed = hash_token(prev_token) % (2**32)
        green, red = partition_vocab(tokenizer.vocab_size, seed, gamma)
        # ç•¶å‰ token æ˜¯å¦åœ¨ç¶ åå–®
        if tokens[i] in green:
            green_count += 1

    # z æª¢é©—
    expected = gamma * T
    var = T * gamma * (1 - gamma)
    if var <= 0:
        return 0, 1.0

    z = (green_count - expected) / np.sqrt(var)
    p_value = 1 - norm.cdf(z) # å–®å°¾æª¢é©—
    return z, p_value
````

````python
TRAINING_WATERMARK_GAMMA = 0.5   # ç¶ åå–®æ¯”ä¾‹
TRAINING_WATERMARK_DELTA = 0.1   # è¨“ç·´æ™‚çš„ logits åç½®ï¼ˆé€šå¸¸æ¯”ç”Ÿæˆæ™‚å°ï¼‰
TRAINING_WATERMARK_HARD = False  # è¨“ç·´æ™‚ä¸å»ºè­°ä½¿ç”¨ç¡¬æ°´å°

# --------------------------
# è¨“ç·´ä¸­ä½¿ç”¨çš„æ°´å° Logits èª¿æ•´
# --------------------------
# æ³¨æ„ï¼šwatermark_sampling å‡½æ•¸éœ€è¦èƒ½å¤ è™•ç† (batch_size, vocab_size) çš„ logits
# ä½†åœ¨è‡ªå®šç¾© Trainer ä¸­ï¼Œæˆ‘å€‘å°‡é€ token è™•ç†ã€‚

# ç¢ºä¿æ‚¨çš„ hash_token å’Œ partition_vocab å·²ç¶“åœ¨è…³æœ¬ä¸­å®šç¾©ã€‚

def apply_watermark_to_training_logits(logits, prev_token_id, gamma, delta, hard=False):
    """
    å°‡æ°´å°åç½®æ‡‰ç”¨æ–¼å–®å€‹ token çš„ logits (ç¶­åº¦: vocab_size)
    """
    vocab_size = logits.shape[-1]

    # æª¢æŸ¥ prev_token_id æ˜¯å¦ç‚ºæœ‰æ•ˆçš„ token ID
    if prev_token_id is None or prev_token_id < 0 or prev_token_id >= vocab_size:
        # å¦‚æœæ˜¯èµ·å§‹ token æˆ– masked tokenï¼Œå‰‡ä¸åŠ æ°´å°
        return logits

    # è¨ˆç®—ç¶ åå–®
    seed = hash_token(prev_token_id) % (2**32)
    green, red = partition_vocab(vocab_size, seed, gamma)

    green_indices = list(green)

    if hard:
        # è¨“ç·´æ™‚å¼·çƒˆä¸å»ºè­°ä½¿ç”¨ç¡¬ç´…åå–®
        raise NotImplementedError("Hard watermarking is not recommended for training loss.")
    else:
        # è»Ÿç´…åå–®ï¼šçµ¦ç¶ è‰²é›†åˆåŠ  Î´ åç½®
        bias = torch.zeros_like(logits, device=logits.device)
        # ä½¿ç”¨ torch.index_put_ ä¾†å®‰å…¨åœ°æ›´æ–°å¤šå€‹ç´¢å¼•
        bias[green_indices] = delta
        logits = logits + bias

    return logits
````

We now have to apply the chat template for `Gemma-3` onto the conversations, and save it to `text`. We remove the `<bos>` token using removeprefix(`'<bos>'`) since we're finetuning. The Processor will add this token before training and the model expects only one.

````python
# Modify the formatting function for the Alpaca dataset
def formatting_prompts_func(examples):
    instructions = examples["instruction"]
    inputs       = examples["input"]
    outputs      = examples["output"]
    texts = []
    for instruction, input, output in zip(instructions, inputs, outputs):
        # Construct the text based on Alpaca format
        if input:
            text = f"<start_of_turn>user\n{instruction}\n{input}<end_of_turn>\n<start_of_turn>model\n{output}<end_of_turn>"
        else:
            text = f"<start_of_turn>user\n{instruction}<end_of_turn>\n<start_of_turn>model\n{output}<end_of_turn>"
        texts.append(text)
    return { "text" : texts, }

dataset = dataset.map(formatting_prompts_func, batched = True)
````

````python
# ----------------------------------------------------
# è³‡æ–™å°é½Šèˆ‡æœ‰æ•ˆæ€§æª¢æŸ¥
# ----------------------------------------------------
def check_dataset_alignment_and_mask(dataset, max_checks=100):
    """
    æª¢æŸ¥è³‡æ–™é›†ä¸­çš„ input_ids å’Œ labels é•·åº¦æ˜¯å¦ä¸€è‡´ï¼Œä¸¦é©—è­‰æ˜¯å¦å­˜åœ¨æœ‰æ•ˆçš„ loss tokenã€‚
    """
    mismatched_count = 0
    empty_label_count = 0
    subset = dataset.select(range(min(len(dataset), max_checks)))
    max_len = model.max_seq_length

    print("-" * 50)
    print(f"æ­£åœ¨å°å‰ {len(subset)} å€‹æ¨£æœ¬é€²è¡Œæ‰¹æ¬¡ Tokenization...")


    # Trainer å…§éƒ¨æœƒå° dataset åŸ·è¡Œ tokenizationï¼Œé€™è£¡æˆ‘å€‘æ‰‹å‹•åŸ·è¡Œä¸€æ¬¡ä¾†æª¢æŸ¥
    # æ³¨æ„: é€™è£¡éœ€è¦ç¢ºä¿åœ¨åŸ·è¡Œå‰ï¼Œæ¨¡å‹çš„ tokenizer å·²ç¶“è¼‰å…¥

    # ç”±æ–¼ SFTTrainer è™•ç†çš„æ˜¯ "text" æ¬„ä½ï¼Œæˆ‘å€‘å¿…é ˆå…ˆä½¿ç”¨ tokenizer è™•ç†
    tokenized_dataset = subset.map(
        # é€™è£¡çš„ x æ˜¯ä¸€å€‹å­—å…¸åˆ—è¡¨ï¼Œx["text"] æ˜¯ä¸€å€‹å­—ç¬¦ä¸²åˆ—è¡¨
        lambda x: tokenizer(x["text"], max_length=max_len, truncation=True, padding=False),
        batched=True, # <-- é€Ÿåº¦æå‡é»
        remove_columns=subset.column_names
    )

    # SFTTrainer é€šå¸¸æœƒè‡ªå‹•å‰µå»º labelsï¼Œä½†æˆ‘å€‘å‡è¨­ tokenized_dataset['input_ids']
    # å·²ç¶“æ˜¯å°é½Šçš„ã€‚é€™è£¡ä¸»è¦æª¢æŸ¥é•·åº¦ä¸€è‡´æ€§ã€‚

    for i in range(min(len(tokenized_dataset), max_checks)):
        sample = tokenized_dataset[i]

        # ç”±æ–¼ SFTTrainer çš„ input_ids æœƒè¢«è¤‡è£½ç‚º labels (ç„¶å¾Œæ‰è¢« mask)
        # æˆ‘å€‘æœŸæœ›å®ƒå€‘åœ¨ tokenizer è¼¸å‡ºæ™‚é•·åº¦æ˜¯ä¸€è‡´çš„
        input_ids = sample["input_ids"]
        # åœ¨ SFTTrainer é‹è¡Œå‰ï¼Œlabels æ¬„ä½å¯èƒ½ä¸å­˜åœ¨ï¼Œé€™è£¡æˆ‘å€‘éœ€è¦å…ˆå‡è¨­å®ƒæ˜¯ input_ids çš„æ‹·è²
        # ä½†ç‚ºäº†é€šé SFTTrainer çš„æµç¨‹ï¼Œæˆ‘å€‘å¿…é ˆç¢ºä¿ input_ids æ˜¯æ­£ç¢ºçš„ã€‚

        # é€™è£¡çš„é—œéµæª¢æŸ¥é»æ˜¯: æ¨¡å‹ tokenizer è™•ç†å¾Œçš„ input_ids æ‡‰è©²æ˜¯æœ‰æ•ˆçš„
        if "labels" in sample:
            labels = sample["labels"]
            if len(input_ids) != len(labels):
                mismatched_count += 1
                print(f"æ¨£æœ¬ {i} è­¦å‘Šï¼šinput_ids ({len(input_ids)}) å’Œ labels ({len(labels)}) é•·åº¦ä¸ä¸€è‡´ã€‚")

            # æª¢æŸ¥æ˜¯å¦å­˜åœ¨æœ‰æ•ˆçš„ loss token
            if all(l == -100 for l in labels):
                empty_label_count += 1
                # é€™æ˜¯å°è‡´ Loss ç‚ºç©ºçš„å¸¸è¦‹åŸå› ï¼
                print(f"æ¨£æœ¬ {i} è­¦å‘Šï¼šæ‰€æœ‰ labels å‡ç‚º -100 (å›æ‡‰éƒ¨åˆ†è¢«å®Œå…¨é®ç½©æˆ–é•·åº¦ç‚º 0)ã€‚")

    print("-" * 50)
    if mismatched_count == 0 and empty_label_count == 0:
        print(f"âœ… å‰ {max_checks} å€‹æ¨£æœ¬çš„ input_ids/labels é•·åº¦ä¸€è‡´ä¸”åŒ…å«æœ‰æ•ˆçš„ Loss tokenã€‚")
    else:
        print(f"âŒ è­¦å‘Šï¼šç™¼ç¾ {mismatched_count} å€‹é•·åº¦ä¸ä¸€è‡´çš„æ¨£æœ¬ï¼Œå’Œ {empty_label_count} å€‹ç„¡æ•ˆ Loss token çš„æ¨£æœ¬ã€‚")
        print("è«‹æª¢æŸ¥æ‚¨çš„ formatting_prompts_func å’Œ train_on_responses_only è¨­ç½®ã€‚")
    print("-" * 50)

# åŸ·è¡Œæª¢æŸ¥ (å°‡æ­¤è¡Œæ”¾åœ¨æ‚¨ Cell 4 çš„ SFTTrainer è¨­ç½®ä¹‹å‰)
check_dataset_alignment_and_mask(dataset, max_checks=100)
````

````python
# ----------------------------------------------------
# æª¢æŸ¥æ¥µçŸ­åºåˆ—é•·åº¦
# ----------------------------------------------------

# è¨­ç½®æ¥µçŸ­é–¾å€¼ (ä¾‹å¦‚ï¼Œå°æ–¼ 5 å€‹ token çš„åºåˆ—å¯èƒ½å°è‡´å•é¡Œ)
SEQUENCE_LENGTH_THRESHOLD = 5
CHECK_SAMPLES_COUNT = 500 # æª¢æŸ¥å‰ 200 å€‹æ¨£æœ¬

def check_short_sequences(dataset, tokenizer, max_len, threshold=5, max_checks=200):
    """
    æª¢æŸ¥ tokenized å¾Œçš„åºåˆ—é•·åº¦æ˜¯å¦éçŸ­ã€‚
    """
    short_sequences_count = 0

    print("-" * 50)
    print(f"æ­£åœ¨æª¢æŸ¥å‰ {max_checks} å€‹æ¨£æœ¬çš„ tokenization é•·åº¦ (é–¾å€¼: {threshold} tokens)...")

    # é¸æ“‡ä¸€å€‹å­é›†é€²è¡Œæª¢æŸ¥
    subset = dataset.select(range(min(len(dataset), max_checks)))

    # é€²è¡Œæ‰¹æ¬¡ tokenization (è¼ƒå¿«)
    tokenized_results = tokenizer(
        subset["text"],
        max_length=max_len,
        truncation=True,
        padding=False,
        return_length=True # è¿”å›é•·åº¦è³‡è¨Š
    )

    lengths = tokenized_results['length']

    for i, length in enumerate(lengths):
        if length <= threshold:
            short_sequences_count += 1
            # ç²å–åŸå§‹æ–‡æœ¬é€²è¡Œå ±å‘Š
            original_text = subset[i]["text"]

            # åªé¡¯ç¤ºæ¨¡å‹çš„å›è¦†éƒ¨åˆ† (é€šå¸¸æ˜¯å›è¦†å¤ªçŸ­å°è‡´çš„å•é¡Œ)
            response_start_tag = "<start_of_turn>model\n"
            try:
                response_start = original_text.rindex(response_start_tag) + len(response_start_tag)
                response_text = original_text[response_start:].replace('<end_of_turn>', '').strip()
            except ValueError:
                response_text = original_text # å¦‚æœæ‰¾ä¸åˆ° tagï¼Œå‰‡é¡¯ç¤ºå…¨éƒ¨

            print(f"æ¨£æœ¬ {i} è­¦å‘Šï¼štoken é•·åº¦éçŸ­ ({length} tokens)ã€‚\n  [å›è¦†] -> {response_text[:80]}...")

    print("-" * 50)
    if short_sequences_count == 0:
        print(f"âœ… å‰ {max_checks} å€‹æ¨£æœ¬çš„ token é•·åº¦å‡ç¬¦åˆè¦æ±‚ (> {threshold} tokens)ã€‚")
    else:
        print(f"âŒ è­¦å‘Šï¼šç™¼ç¾ {short_sequences_count} å€‹ token é•·åº¦éçŸ­çš„æ¨£æœ¬ã€‚")
        print("  å»ºè­°æª¢æŸ¥é€™äº›æ¨£æœ¬çš„åŸå§‹å›è¦†å…§å®¹æ˜¯å¦éæ–¼ç°¡çŸ­ã€‚")
    print("-" * 50)


# ç¢ºä¿ model.max_seq_length å¯ç”¨
max_len = model.max_seq_length

# åŸ·è¡Œæª¢æŸ¥
check_short_sequences(dataset, tokenizer, max_len, threshold=5, max_checks=CHECK_SAMPLES_COUNT)
````

````python
from trl import SFTConfig,SFTTrainer
import torch.nn.functional as F # ç¢ºä¿åœ¨æ–‡ä»¶é–‹é ­å°å…¥
class WatermarkSFTTrainer(SFTTrainer):
    """
    è‡ªå®šç¾© SFTTrainerï¼Œåœ¨è¨ˆç®— Loss æ™‚åŠ å…¥ç´…ç¶ åå–®æ°´å°çš„ Logits åç½®
    """
    def __init__(self, *args, watermark_gamma=0.5, watermark_delta=0.1, **kwargs):
        super().__init__(*args, **kwargs)
        self.watermark_gamma = watermark_gamma
        self.watermark_delta = watermark_delta

    def compute_loss(self, model, inputs, return_outputs=False,**kwargs):
        # 1. è™•ç† labels: ç”±æ–¼ train_on_responses_only å·²è¨»è§£ï¼Œé€™è£¡ä¸éœ€è¦ pop labels
        # ç¢ºä¿ inputs ä¸­æœ‰ labels æ¬„ä½
        labels = inputs.get("labels")
        if labels is None:
            # SFTTrainer åœ¨æ­¤æ™‚æ‡‰è©²å·²ç¶“å‰µå»ºäº† labelsï¼Œå¦‚æœæ²’æœ‰ï¼Œç„¡æ³•ç¹¼çºŒ
            print("åš´é‡è­¦å‘Šï¼šinputs ä¸­ç¼ºå°‘ 'labels' æ¬„ä½ã€‚")
            loss = torch.tensor(0.0, device=model.device, requires_grad=True)
            return (loss, outputs) if return_outputs else loss

        # 2. å‰å‘å‚³æ’­
        outputs = model(**inputs, return_dict=True)

        # 3. ç²å– Logits (ä½¿ç”¨ä¿è­·æ©Ÿåˆ¶)
        try:
            logits = outputs.to_tuple()[0]
        except Exception:
            logits = outputs.logits

        # 4. Logits ç©ºæª¢æŸ¥ (ä¿æŒä¸å‹•ï¼Œä½œç‚ºé˜²è­·)
        if logits is None or logits.dim() < 3 or logits.shape[0] == 0 or logits.shape[1] == 0:
            print("è­¦å‘Šï¼šæ•ç²åˆ°ç©º Logits æ‰¹æ¬¡ï¼Œè·³éè¨ˆç®—ã€‚")
            loss = torch.tensor(0.0, device=model.device, requires_grad=True)
            return (loss, outputs) if return_outputs else loss

        B, L, V = logits.shape

        # 5. æ‡‰ç”¨æ°´å°åˆ° Logits (ä½ çš„æ°´å°é‚è¼¯)
        adjusted_logits = torch.empty_like(logits)
        # ... (é€ token æ‡‰ç”¨ adjusted_logits çš„é‚è¼¯ - æ­¤è™•éœ€è¦è‡ªè¡Œå¯¦ç¾æˆ–è¤‡è£½éä¾†) ...
        # (é€™è£¡å‡è¨­ä½ çš„é€ token æ°´å°é‚è¼¯å·²ç¶“å¯¦ä½œä¸¦è³¦å€¼çµ¦ adjusted_logits)

        # ç‚ºäº†æ¼”ç¤º Loss è¨ˆç®—ï¼Œæˆ‘å€‘å‡è¨­ adjusted_logits å·²ç¶“è¢«è¨ˆç®—å‡ºä¾†

        # 6. !!! æ ¸å¿ƒä¿®æ­£ï¼šç›´æ¥ä½¿ç”¨ PyTorch è¨ˆç®— Loss !!!

        # a. ç¢ºä¿ logits å’Œ labels ç¶­åº¦æ­£ç¢ºä»¥è¨ˆç®— CrossEntropyLoss
        # Logits: [B*L, V]
        shift_logits = adjusted_logits[..., :-1, :].contiguous()
        # Labels: [B*L]
        shift_labels = labels[..., 1:].contiguous()

        # b. èª¿æ•´ç¶­åº¦ä»¥é©æ‡‰ F.cross_entropy
        loss = F.cross_entropy(
            shift_logits.view(-1, V),
            shift_labels.view(-1),
            # ç¢ºä¿å¿½ç•¥ç´¢å¼•èˆ‡æœªè¢« mask çš„ token ID -100 ä¸€è‡´
            ignore_index=-100
        )

        # 7. è¿”å› Loss
        return (loss, outputs) if return_outputs else loss
````

````python
def detect_watermark(text, tokenizer, gamma=0.5):
    """
    å°æ–‡æœ¬åŸ·è¡Œ Z æª¢é©—ä»¥æª¢æ¸¬ç¶ åå–®æµ®æ°´å°

    Args:
        text (str): å¾…æª¢æ¸¬çš„æ–‡æœ¬ã€‚
        tokenizer: ç”¨æ–¼å°‡æ–‡æœ¬è½‰æ›ç‚º token ID çš„ tokenizerã€‚
        gamma (float): æµ®æ°´å°ç”Ÿæˆæ™‚ä½¿ç”¨çš„ç¶ åå–®æ¯”ä¾‹ (é æœŸæ©Ÿç‡)ã€‚

    Returns:
        tuple: (z_score, p_value)
    """

    # 1. Tokenization: ç¢ºä¿ä¸æ·»åŠ é¡å¤–çš„ç‰¹æ®Š tokenï¼Œåªç²å–å…§å®¹ token
    # å¿½ç•¥ text_pair=False
    tokens = tokenizer(text, return_tensors="pt", add_special_tokens=False).input_ids[0].tolist()

    if not tokens:
        print("è­¦å‘Šï¼šæ–‡æœ¬ç„¡æ³• tokenization æˆ–ç‚ºç©ºã€‚")
        return 0, 1.0

    # æª¢æ¸¬æ˜¯å¾ç¬¬äºŒå€‹ token (i=1) é–‹å§‹çš„ï¼Œå› ç‚ºç¬¬ä¸€å€‹ token æ²’æœ‰å‰å° token ä¾†è¨ˆç®—ç¶ åå–®ã€‚
    # T æ˜¯å¯æª¢æ¸¬çš„ token æ•¸é‡ï¼Œå³ï¼šåºåˆ—é•·åº¦ - 1ã€‚
    T = len(tokens) - 1
    if T <= 0:
        print("è­¦å‘Šï¼šæ–‡æœ¬å¤ªçŸ­ (<= 1 token)ï¼Œç„¡æ³•åŸ·è¡Œæª¢æ¸¬ã€‚")
        return 0, 1.0

    green_count = 0

    print(f"ç¸½å…± {len(tokens)} å€‹ tokenï¼Œå…¶ä¸­ {T} å€‹ token å¯ç”¨æ–¼æª¢æ¸¬ã€‚")

    for i in range(1, len(tokens)):
        # å‰ä¸€å€‹ token ç”¨æ–¼ç”Ÿæˆç¨®å­
        prev_token = tokens[i - 1]
        current_token = tokens[i]

        # ç²å–ç¶ åå–®
        seed = hash_token(prev_token) % (2**32)
        green, red = partition_vocab(tokenizer.vocab_size, seed, gamma)

        # ç•¶å‰ token æ˜¯å¦åœ¨ç¶ åå–®
        if current_token in green:
            green_count += 1

    # 2. çµ±è¨ˆ Z æª¢é©—
    # é›¶å‡è¨­ H0: token éš¨æ©Ÿç”Ÿæˆ (è½åœ¨ç¶ åå–®çš„æ©Ÿç‡ = gamma)
    # å‚™æ“‡å‡è¨­ H1: token æœ‰ç¶ åå–®åç½® (è½åœ¨ç¶ åå–®çš„æ©Ÿç‡ > gamma)

    expected = gamma * T
    var = T * gamma * (1 - gamma)

    if var <= 1e-6: # é¿å…é™¤ä»¥é›¶æˆ–æ¥µå°çš„æ•¸
        print("è­¦å‘Šï¼šè®Šç•°æ•¸éå°ã€‚")
        return 0, 1.0

    # è¨ˆç®— Z-score
    z = (green_count - expected) / np.sqrt(var)

    # è¨ˆç®— p-value (å–®å°¾æª¢é©—ï¼Œå› ç‚ºæˆ‘å€‘åªé—œå¿ƒç¶ åå–®æ•¸é‡æ˜¯å¦éå¤š)
    p_value = 1 - norm.cdf(z)

    print("-" * 30)
    print(f"ç¶ åå–®æ¯”ä¾‹ (Î³): {gamma}")
    print(f"è§€å¯Ÿåˆ°çš„ç¶ åå–®æ•¸é‡: {green_count} / {T}")
    print(f"é æœŸçš„ç¶ åå–®æ•¸é‡ (H0): {expected:.2f}")
    print("-" * 30)

    return z, p_value

print("æª¢æ¸¬å·¥å…·å‡½å¼å®šç¾©å®Œæˆã€‚")
````
